```python
from unsloth import FastLanguageModel
from peft import PeftModel
import torch

# Load Llama 3 8B base model with 4-bit quantization
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/llama-3-8b-bnb-4bit",
    max_seq_length = 8192,
    dtype = torch.float16,
    load_in_4bit = True,
)

# Load Centaur 8B adapter
model = PeftModel.from_pretrained(
    model,
    "marcelbinz/Llama-3.1-Centaur-8B-adapter",
    adapter_name = "centaur"
)

# Merge adapter into base model for inference
model.merge_and_unload()
```

```python
prompt = "Explain what Centaur mode is in simple terms."
inputs = tokenizer(prompt, return_tensors="pt").to("cuda")

outputs = model.generate(
    **inputs,
    max_new_tokens=200,
    temperature=0.7,
    do_sample=True,
)

print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```
